1. What is a random variable in probability theory?
  - A random variable is a function that maps each outcome in a sample space to a real number.

  Why we need random variables

In probability, outcomes are often non-numerical (like Heads/Tails or rain/no rain). Random variables let us work with numbers, making it possible to calculate probabilities, averages, variance, etc.

Example 1: Coin toss

Experiment: Toss a coin

Sample space: {H, T}

Define a random variable X as:

X = 1 if Head occurs

X = 0 if Tail occurs

Here, X is the random variable.

3.Trials are independent

4.Probability of success (p) is constant for every trial

6.  What is the Central Limit Theorem (CLT), and why is it critical in statistics?
  - What is the Central Limit Theorem?

The CLT states that:

For a large enough sample size, the distribution of the sample mean becomes approximately normal, regardless of the shape of the original population distribution, provided the population has a finite mean and variance.

Why is CLT critical in statistics?
1. Makes statistical inference possible

It allows us to use normal distribution methods (Z-tests, confidence intervals) even when the population is not normal.

2. Foundation of hypothesis testing

Most tests rely on the assumption that the sampling distribution of the mean is normal ‚Äî CLT justifies this.

3. Works with real-world data

Real data is often skewed or irregular. CLT lets us still analyze it reliably.

4. Explains accuracy with larger samples

As n increases:

* Sampling variability decreases
* Estimates become more precise

7. What is the significance of confidence intervals in statistical analysis?
  - What is a confidence interval?

A confidence interval is a range of values calculated from sample data that is likely to contain the true population parameter (such as a mean or proportion).

Example:
A 95% confidence interval for the mean income might be:
(45,000,50,000)

Why are confidence intervals significant?
1. Show uncertainty (not just a single number)

A point estimate (like a sample mean) hides uncertainty.
A CI shows how reliable that estimate is.

2. More informative than hypothesis tests

Instead of just ‚Äúreject‚Äù or ‚Äúfail to reject,‚Äù CIs:

* Show direction
* Show magnitude
* Show precision

3. Help in decision-making

* Narrow CI ‚Üí precise estimate
* Wide CI ‚Üí more data or better sampling needed

Used in medicine, economics, business, and social sciences.

4. Connect directly to sample size and variability

Larger sample size ‚Üí narrower CI

Higher variability ‚Üí wider CI

Higher confidence level (99% vs 95%) ‚Üí wider CI

5. Useful for hypothesis testing

If a CI for a mean difference does not include 0, the result is statistically significant (at that confidence level).

It tells what you can expect in the long run, not in a single trial.

2. Foundation of decision-making

```python
#9. Write a Python program to generate 1000 random numbers from a normal
# distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
# deviation using NumPy, and draw a histogram to visualize the distribution.

import numpy as np
import matplotlib.pyplot as plt

# Parameters
mean = 50
std_dev = 5
n = 1000

# Generate 1000 random numbers from a normal distribution
data = np.random.normal(mean, std_dev, n)

# Compute mean and standard deviation
computed_mean = np.mean(data)
computed_std = np.std(data)

print("Computed Mean:", computed_mean)
print("Computed Standard Deviation:", computed_std)

# Plot histogram
plt.hist(data, bins=30)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Normally Distributed Random Numbers")
plt.show()
```

How to apply the Central Limit Theorem (CLT) to estimate average sales
Step 1: Define the goal

The company wants to estimate the true average daily sales (population mean) using sample data and provide a 95% confidence interval.

Step 2: Use the Central Limit Theorem

Even if daily sales are not perfectly normally distributed, the CLT tells us that:

For a sufficiently large sample size, the sampling distribution of the sample mean will be approximately normal.

This allows us to use normal (Z) statistics to estimate the average sales.

Step 3: Compute sample statistics

From the given daily sales data:

* Sample mean (xÀâ) ‚âà 249.25
* Sample standard deviation (s) ‚âà 17.53
* Sample size (n) = 20

Step 4: Calculate the standard error
Standard Error (SE)=ùë†/ùëõ=17.53/root 20‚âà3.92

Step 5: Construct the 95% confidence interval

For a 95% confidence level:

Step 6: Final 95% confidence interval

* Understand expected daily sales

```python
## Write the Python code to compute the mean sales and its confidence interval.
import numpy as np
import math

# Daily sales data
daily_sales = [
    220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
    235, 260, 245, 250, 225, 270, 265, 255, 250, 260
]

# Convert to NumPy array
sales = np.array(daily_sales)

# Compute sample statistics
mean_sales = np.mean(sales)
std_sales = np.std(sales, ddof=1)   # sample standard deviation
n = len(sales)

# Standard error
standard_error = std_sales / math.sqrt(n)

# Z-value for 95% confidence interval
z = 1.96

# Confidence interval
lower_bound = mean_sales - z * standard_error
upper_bound = mean_sales + z * standard_error

print("Mean Daily Sales:", mean_sales)
print("95% Confidence Interval:", (lower_bound, upper_bound))
```
