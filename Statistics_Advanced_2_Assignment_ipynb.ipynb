1.  What is hypothesis testing in statistics?
  - Hypothesis testing is a formal way to make decisions using data. You start with a claim about a population, collect a sample, and then use probability to decide whether the data supports that claim or not.

Think of it as: “Is what I'm seeing in the data just random chance, or is something real going on?”

The basic idea

You set up two competing hypotheses:

**Null hypothesis (H₀)**
This is the default assumption. It usually says nothing has changed or there is no effect.
Example:

H₀: The average score is 50.

**Alternative hypothesis (H₁ or Hₐ)**
This is what you want to test for. It represents a change or effect.
Example:

H₁: The average score is not 50.

2.  What is the null hypothesis, and how does it differ from the alternative
hypothesis?
  -** Null Hypothesis (H₀)**

The null hypothesis is the default assumption. It states that there is no effect, no difference, or no relationship in the population.

It represents the status quo

Any observed difference is assumed to be due to random chance

Examples:

H₀: The average marks of students = 70

H₀: A new medicine has no effect

H₀: There is no difference between two machine outputs

**Alternative Hypothesis (H₁ or Hₐ)**

The alternative hypothesis is the opposite of the null hypothesis. It states that there is an effect, difference, or relationship.

It's what the researcher wants to find evidence for

It challenges the null hypothesis

Examples:

H₁: The average marks of students ≠ 70

H₁: A new medicine has an effect

H₁: There is a difference between two machine outputs

3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
  - The significance level, denoted by α (alpha), is a pre-chosen threshold that helps you decide whether the result of a hypothesis test is statistically meaningful or just due to random chance.

What is the significance level (α)?

It is the maximum probability of making a Type I error.

A Type I error occurs when you reject a true null hypothesis.

In simple terms:

α tells you how much risk you’re willing to take of making a wrong rejection.

Common values:

α = 0.05 (5%) → most common

α = 0.01 (1%) → stricter tests

α = 0.10 (10%) → more lenient tests

Role of α in hypothesis testing

Set before analyzing data
You choose α before doing the test to avoid bias.

Acts as a decision boundary
You compare the p-value with α:

If p-value ≤ α → Reject H₀

If p-value > α → Fail to reject H₀

Defines the rejection region
α determines the critical value(s) on the test distribution (z, t, etc.) beyond which H₀ is rejected.

Example

Suppose:

α = 0.05

p-value = 0.03

Since 0.03 < 0.05, the result is statistically significant → reject the null hypothesis.

If instead:

p-value = 0.08

Then 0.08 > 0.05 → fail to reject the null hypothesis.

4.  What are Type I and Type II errors? Give examples of each.
  - Type I Error (False Positive)
What it is

A Type I error happens when you reject the null hypothesis even though it is actually true.

Symbol: α (alpha)

In simple words: You think there is an effect, but there isn’t.

Example

Medical test example:

H₀: A person does not have a disease

Decision: Test says the person has the disease

Reality: Person is actually healthy

Type II Error (False Negative)
What it is

A Type II error happens when you fail to reject the null hypothesis even though it is false.

Symbol: β (beta)

In simple words: You miss a real effect.

Example

Medical test example:

H₀: A person does not have a disease

Decision: Test says the person is healthy

Reality: Person actually has the disease

5.  What is the difference between a Z-test and a T-test? Explain when to use
each.
  - The Z-test and T-test are both hypothesis tests used to compare means, but they differ mainly in what you know about the population and sample size.

  Z-test
What it is

A Z-test uses the standard normal distribution (Z distribution).

When to use a Z-test

Use a Z-test when:

The population standard deviation (σ) is known

The sample size is large (typically n ≥ 30)

Data is approximately normally distributed (or n is large enough by CLT)

Example

Testing whether the average height of students is 165 cm

σ = 10 cm (known)

n = 50

 Use a Z-test

T-test
What it is

A T-test uses the t-distribution, which accounts for extra uncertainty when σ is unknown.

When to use a T-test

Use a T-test when:

The population standard deviation is unknown

The sample size is small (n < 30)

Data is approximately normally distributed

Example

Testing whether the average exam score is 70

σ is unknown

n = 12

 Use a T-test

```python
''' 6. Write a Python program to generate a binomial distribution with n=10 and
 p=0.5, then plot its histogram.'''

import random
import matplotlib.pyplot as plt

# Parameters
n = 10        # number of trials
p = 0.5       # probability of success
samples = 1000  # number of experiments

# Generate binomial data
binomial_data = []

for _ in range(samples):
    successes = 0
    for _ in range(n):
        if random.random() < p:
            successes += 1
    binomial_data.append(successes)

# Plot histogram
plt.hist(binomial_data, bins=range(0, n + 2))
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.show()
```

```python
''' 7.  Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
                        50.3, 50.4, 50.0, 49.7, 50.5, 49.9] '''

import math
from statistics import mean, stdev
from scipy.stats import norm

# Sample data
sample_data = [
    49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
    50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
    50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
    50.3, 50.4, 50.0, 49.7, 50.5, 49.9
]

# Hypothesized population mean
mu_0 = 50
alpha = 0.05

# Sample statistics
x_bar = mean(sample_data)
sigma = stdev(sample_data)     # sample std deviation
n = len(sample_data)

# Z-statistic
z = (x_bar - mu_0) / (sigma / math.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

print("Sample Mean:", round(x_bar, 4))
print("Z-statistic:", round(z, 4))
print("P-value:", round(p_value, 4))
```

```python
''' 8.  Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib. '''

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data from a normal distribution
np.random.seed(42)      # for reproducibility
mu = 50                 # true mean
sigma = 5               # standard deviation
n = 100                 # sample size

data = np.random.normal(mu, sigma, n)

# Step 2: Calculate 95% confidence interval for the mean
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)

confidence_level = 0.95
alpha = 1 - confidence_level

z_critical = stats.norm.ppf(1 - alpha / 2)
margin_of_error = z_critical * (sample_std / np.sqrt(n))

ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print("Sample Mean:", round(sample_mean, 3))
print("95% Confidence Interval: (", round(ci_lower, 3), ",", round(ci_upper, 3), ")")

# Step 3: Plot the data
plt.hist(data, bins=15)
plt.axvline(sample_mean)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Simulated Normal Distribution Data")
plt.show()
```

```python
''' 9. Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.'''

import numpy as np
import matplotlib.pyplot as plt

# Sample dataset
data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5]

# Function to calculate Z-scores
def calculate_z_scores(dataset):
    mean = np.mean(dataset)
    std_dev = np.std(dataset, ddof=1)   # sample standard deviation
    z_scores = [(x - mean) / std_dev for x in dataset]
    return z_scores

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Plot histogram of standardized data
plt.hist(z_scores, bins=10)
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.title("Histogram of Z-scores (Standardized Data)")
plt.show()

# Display some Z-scores
print("First 5 Z-scores:", [round(z, 3) for z in z_scores[:5]])
```

Explanation: What Z-scores Represent

A Z-score tells you:

How many standard deviations a data point is away from the mean

The formula is:
                  Z=x-μ​/σ

Interpretation

Z = 0 → value is exactly at the mean

Z = +1 → value is 1 standard deviation above the mean

Z = -2 → value is 2 standard deviations below the mean

Example from output

Z ≈ -2.10 → value is far below the mean
